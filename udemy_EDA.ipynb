### Visualize the count of subjects available in this dataset 

```python
plt.figure(figsize=(15,8))
sb.countplot(x='subject',data=df)
```

Here we can see we have more courses on business finance and web development than courses on musincal instrument and graphc design

### Count the number of free and paid courses

```python
df.is_paid.unique()
```

```python
num_paid = len(df[df.is_paid==True])
num_free = len(df[df['is_paid']==False])
print('Number of Free courses = {}\n Number of Paid courses= {}'.format(num_free,num_paid))
```

##### lets visualize count of free and paid couses

```python
sb.countplot(x='is_paid',data=df)
```

### Top 10 free courses and top 10 paid courses

```python
df[df.is_paid==True].sort_values('num_subscribers',ascending=False).head(10)[['course_title','num_subscribers']].reset_index()
```

```python
df[df.is_paid==False].sort_values('num_subscribers',ascending=False).head(10)[['course_title','num_subscribers']].reset_index()
```
